```python
         # Формирование матриц
Matrix_A = [[2, -4], [3, 5], [-1, 0]]
Matrix_B = [[1, 2, 7], [-3, -4, 0], [5, 2, 1]]
Matrix_C = [[6, -3, 9], [4, -5, 2], [8, 1, 5]]

         # Вариант [1.1]-1
         # Транспонирование матриц
Matrix_A_T = [[Matrix_A[j][i] for j in range(len(Matrix_A))] for i in range(len(Matrix_A[0]))] 
Matrix_B_T = [[Matrix_B[j][i] for j in range(len(Matrix_B))] for i in range(len(Matrix_B[0]))] 
         # Перемножение матриц (формирование слагаемых)
Matrix_D_1_2 = [[sum(Matrix_A_T[i][k] * Matrix_C[k][j] for k in range(len(Matrix_A_T[0]))) for j in range(len(Matrix_C[0]))] for i in range(len(Matrix_A_T))]
Matrix_D_2_2 = [[(-2)*sum(Matrix_A_T[i][k] * Matrix_B_T[k][j] for k in range(len(Matrix_A_T[0]))) for j in range(len(Matrix_B_T[0]))] for i in range(len(Matrix_A_T))]
         # Формирование и печать результата 
Matrix_D = [[Matrix_D_1_2[i][j] + Matrix_D_2_2[i][j] for j in range(len(Matrix_D_1_2[0]))] for i in range(len(Matrix_D_1_2))]
print("Вариант 1:\n", Matrix_D)

         # Вариант [1.1]-2
Arr_A = np.array(Matrix_A)
Arr_B = np.array(Matrix_B)
Arr_C = np.array(Matrix_C)
         # Формирование результата 
print("Вариант 2:\n", Arr_A.T @ Arr_C - 2 * Arr_A.T @ Arr_B.T)
```

```python
         # Определение исходных параметров
n = 128
Matrix_A = np.array(np.random.randint(1000, size=(n, n)))
Ranks = []
Errors = []
         # Сингулярное разложение матрицы
U, S, Vh = np.linalg.svd(Matrix_A)
print (Matrix_A)
         # Аппроксимация матрицы с рангом от 2 до n и расчет ошибки для каждого ранга
for i in range(2, n+1):
    Matrix_A_Approx = U[:, :i] @ np.diag(S[:i]) @ Vh[:i, :]
    DifferencesMatrix = Matrix_A - Matrix_A_Approx
    Ranks.append(i)
    Errors.append(round(np.sqrt(np.sum(np.square(DifferencesMatrix))),3))

         # Построение графика зависимости ошибки аппроксимации от ранга
plt.plot(Ranks, Errors)  
plt.xlabel("Ранг")  
plt.ylabel("Ошибка аппроксимации")  
plt.title("График зависимости ошибки аппроксимации от ранга")  
plt.show() 
```
